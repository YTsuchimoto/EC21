```python
from sympy import *
from sympy.abc import *
from sympy import pi
import sympy
t = sympy.Symbol('t',real=True)
PAx = 0
PAy = 2

P1x = 5*sympy.sin(2*t)
P1y = 5*sympy.cos(2*t)

P2x = 10*1*sympy.sin(t)
P2y = 10*1*sympy.cos(t)

P1 = sympy.Matrix([P1x-PAx,P1y-PAy,0])
P2 = sympy.Matrix([P2x-PAx,P2y-PAy,0])
print(P1)

S = P1.cross(P2).norm()

print(S)
diff = sympy.diff(S)
print(diff)
plot(S,(t,0,2*pi))
plot(diff,(t,0,2*pi))


#ｔの範囲を(pi/2,pi)にする
S2 = 5*(10*cos(t) - 2)*sin(2*t) - 10*(5*cos(2*t) - 2)*sin(t)
diff2 = sympy.diff(S2)
print(diff2)
plot(S2,(t,0,2*pi))
plot(diff2,(t,0,2*pi))

res = sympy.solve(diff2)
print(res)
Smax = S2.subs(t,res[0])
print(Smax)








```
